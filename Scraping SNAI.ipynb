```python
import pprint
detail_quote = []
for i in range(0, 66, 2):
  Team = wd.find_elements(by=By.CSS_SELECTOR, value="div.col-xs-6.padding-5.text-center")[i].text
  Quota = wd.find_elements(by=By.CSS_SELECTOR, value="div.col-xs-6.padding-5.text-center")[i+1].text

  detail_quote.append({'Quota': Quota,
                       'Team': Team})
  i = i+2

len(detail_quote)
pprint.pprint(detail_quote[0:10])
```

```python
import pandas as pd
detail_quote = pd.DataFrame(detail_quote)
detail_quote.set_index("Team")
detail_quote.head()
detail_quote.to_csv('snai.csv')
```
